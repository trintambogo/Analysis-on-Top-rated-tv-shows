# This is a dataset from kaggle that analyzes the popularity of shows among different countries. My aim was to see how popularity has grown over the years and in which countries are shows most popular

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('data_TV.csv')
```

```python
#first_air_date- the first day the tv show was aired
#origin_country- the country where the tv show was created
#origin_language- the origin language of the show
# name- name of the show
#popularity- a metric that measures how popular a tv show is based on popularity
#number of views in a day,marked as favorite,added to their watchlist
#vote_average - average total number of votes the show received
#vote_count - the number of votes the show received
#overview - a brief description of the show
```

```python
df.info()
#data contains 8 columns and 2543 rows
```

```python
df.first_air_date.fillna('0',inplace=True)
```

```python
df['first_air_date'].isna().sum()
```

```python
df.info()
```

```python
df.tail()
```

```python
df[df['popularity']==df['popularity'].max()]
#house of dragon was the most popular show
```

```python
df[df['popularity']==df['popularity'].min()]
#kungfu was the least popular tv show
```

```python
df[df['vote_count']==df['vote_count'].max()]
#game of thrones received the most number of votes
```

```python
by_country=df.groupby('origin_country').sum()
```

```python
by_country
```

```python
df[df['origin_country']=='c("CA", "US")']
```

```python
df=df.drop(df.index[43])
```

```python
by_country
```

```python
#data visualizations
```

```python
by_country=by_country.reset_index()
```

```python
by_country['popularity'].plot(kind='barh',color='g', figsize=(8,8))
#movies that originate from the US are most popular
#movies that originate in argentina,turkey,br received least popularity
```

```python
df.head()
```

```python
pop=df[df['popularity']>=1000]
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
plt.xticks(rotation='vertical')
sns.barplot(x='name',y='popularity',data=pop)
```

```python
pop=df[df['popularity']<5]
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
plt.xticks(rotation='vertical')
sns.barplot(x='name',y='popularity',data=pop)
#kungfu is the least popular
```
